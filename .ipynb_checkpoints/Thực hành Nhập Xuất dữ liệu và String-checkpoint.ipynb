## Bài 3
Nhập vào `tien_luong_theo_gio` và `so_gio` trong 1 tháng, và `tien_thuong`. Sau đo in ra tổng thu nhập trong tháng.

biết rằng: Tổng thu nhập = tiền lương theo giờ * số giờ + tiền thường

```python
tien_luong_theo_gio = float(input("Tiền lương theo giờ:"))
so_gio = float(input("Số giờ:"))
tien_thuong = float(input("Tiền thưởng:"))
tong_thu_nhap = (tien_luong_theo_gio*so_gio)+tien_thuong
print(f"Tổng thu nhập là {tong_thu_nhap}")
```

## Bài 4

nhập vào một `string` và một số `x`. Hãy xoá `x` ký tự đầu tiên của `string`
Ví dụ: string = 'Hello Python', x=3
Output:'lo Python'

```python
string = input("Nhap mot chuoi bat ky: ")
x = int(input("Nhập x:"))
string = string[x:]
print(f'Chuỗi ký tự sau khi xoá: {string}')
```

```python
from tkinter import *
from tkinter import ttk
import mysql.connector
from tkinter import messagebox
from datetime import datetime

class Employee:
    def __init__(self,root):
        self.root=root
        self.root.geometry("1530x790")
        self.root.title('Employee Management System')
        
        #Variables
        self.var_dep=StringVar()
        self.var_designition=StringVar()
        self.var_doj=StringVar()
        self.var_salary=StringVar()
        self.var_idemployee=StringVar()
        self.var_name=StringVar()
        self.var_gender=StringVar()
        self.var_married=StringVar()
        self.var_idproofcomb=StringVar()
        self.var_idproof=StringVar()
        self.var_country=StringVar()
        self.var_email=StringVar()
        self.var_phone=StringVar()
        self.var_dob=StringVar()
        self.var_address=StringVar()
        self.var_seniority=StringVar()
        self.var_expbonus=StringVar()
        
        lbl_title=Label(self.root,text="EMPLOYEE MANAGEMENT SYSTEM",font=('time new roman',37,'bold'),fg='darkblue',bg='white')
        lbl_title.place(x=0,y=0,width=1410,height=50)
        #Main Frame
        main_frame=Frame(self.root,bd=2,relief=RIDGE,bg='white')
        main_frame.place(x=10,y=70,width=1345,height=615)
        #upper Frame
        upper_frame=LabelFrame(main_frame,bd=2,relief=RIDGE,text='Employee Information',font=('time new roman',13,'bold'),fg='red',bg='white')
        upper_frame.place(x=10,y=10,width=1320,height=290)
        
        #labels and entr fields
        #Label Department
        lbl_dep=Label(upper_frame,text='Department:',font=('arial',12,'bold'),bg='white')
        lbl_dep.grid(row=0,column=0,padx=2,sticky=W)
        
        combo_dep=ttk.Combobox(upper_frame,textvariable=self.var_dep,font=('arial',12),width=20,state='readonly')
        combo_dep['value']=('Select Department','HR','Software Engineer','Manager')
        combo_dep.current(0)
        combo_dep.grid(row=0,column=1,padx=2,pady=7,sticky=W)
        
        #Label designition
        lbl_designition=Label(upper_frame,text='Designition:',font=('arial',12,'bold'),bg='white')
        lbl_designition.grid(row=1,column=0,padx=2,pady=7,sticky=W)
        
        txt_designition=ttk.Entry(upper_frame,textvariable=self.var_designition,font=('arial',12),width=22)
        txt_designition.grid(row=1,column=1,padx=2,pady=7,sticky=W)
        
        #Label doj (Date Of Joining)
        lbl_doj=Label(upper_frame,text='DOJ:',font=('arial',12,'bold'),bg='white')
        lbl_doj.grid(row=2,column=0,padx=2,pady=7,sticky=W)
        
        txt_doj=ttk.Entry(upper_frame,textvariable=self.var_doj,font=('arial',12),width=22)
        txt_doj.grid(row=2,column=1,padx=2,pady=7,sticky=W)
        
        #Label CTC (salary)
        lbl_ctc=Label(upper_frame,text='Salary (USD):',font=('arial',12,'bold'),bg='white')
        lbl_ctc.grid(row=3,column=0,padx=2,pady=7,sticky=W)
        
        txt_ctc=ttk.Entry(upper_frame,textvariable=self.var_salary,font=('arial',12),width=22)
        txt_ctc.grid(row=3,column=1,padx=2,pady=7,sticky=W)
        
        #Label id employee
        lbl_idemployee=Label(upper_frame,text='ID Employee:',font=('arial',12,'bold'),bg='white')
        lbl_idemployee.grid(row=4,column=0,padx=2,pady=7,sticky=W)
        
        txt_idemployee=ttk.Entry(upper_frame,textvariable=self.var_idemployee,font=('arial',12),width=22)
        txt_idemployee.grid(row=4,column=1,padx=1,pady=7)
        
        #Label name
        lbl_name=Label(upper_frame,text='Name:',font=('arial',12,'bold'),bg='white')
        lbl_name.grid(row=0,column=2,padx=2,pady=7,sticky=W)
        
        txt_name=ttk.Entry(upper_frame,textvariable=self.var_name,font=('arial',12),width=22)
        txt_name.grid(row=0,column=3,padx=1,pady=7)
        
        #label Gender
        lbl_gender=Label(upper_frame,text='Gender:',font=('arial',12,'bold'),bg='white')
        lbl_gender.grid(row=1,column=2,padx=2,pady=7,sticky=W)
        
        combo_txt_gender=ttk.Combobox(upper_frame,textvariable=self.var_gender,font=('arial',12),width=20,state='readonly')
        combo_txt_gender['value']=('Male','Female','Other')
        combo_txt_gender.current(0)
        combo_txt_gender.grid(row=1,column=3,padx=2,pady=7,sticky=W)
        
        #Label married
        lbl_married_status=Label(upper_frame,text='Married Status:',font=('arial',12,'bold'),bg='white')
        lbl_married_status.grid(row=2,column=2,padx=2,pady=7,sticky=W)
        
        combo_txt_married=ttk.Combobox(upper_frame,textvariable=self.var_married,font=('arial',12),width=20,state='readonly')
        combo_txt_married['value']=('Married','Unmarried')
        combo_txt_married.current(0)
        combo_txt_married.grid(row=2,column=3,padx=2,pady=7,sticky=W)
        
        #label ID
        combo_txt_id=ttk.Combobox(upper_frame,textvariable=self.var_idproofcomb,font=('arial',12,'bold'),width=17,state='readonly')
        combo_txt_id['value']=('Select  ID Proof','IDENTITY CARD:','CITIZENSHIP CARD:','PASSPORT','DRIVING LICENCE:')
        combo_txt_id.current(0)
        combo_txt_id.grid(row=3,column=2,padx=2,pady=7,sticky=W)
        
        txt_id=ttk.Entry(upper_frame,textvariable=self.var_idproof,font=('arial',12),width=22)
        txt_id.grid(row=3,column=3,padx=2,pady=7,sticky=W)
        
        #Label Country
        lbl_country=Label(upper_frame,text='Country:',font=('arial',12,'bold'),bg='white')
        lbl_country.grid(row=4,column=2,padx=2,pady=7,sticky=W)
        
        txt_country=ttk.Entry(upper_frame,textvariable=self.var_country,font=('arial',12),width=22)
        txt_country.grid(row=4,column=3,padx=2,pady=7,sticky=W)
        
        #Label email
        lbl_email=Label(upper_frame,text='Email:',font=('arial',12,'bold'),bg='white')
        lbl_email.grid(row=0,column=4,padx=2,pady=7,sticky=W)
        
        txt_email=ttk.Entry(upper_frame,textvariable=self.var_email,font=('arial',12),width=22)
        txt_email.grid(row=0,column=5,padx=2,pady=7)
        
        #Label Phone
        lbl_phone=Label(upper_frame,text='Phone No:',font=('arial',12,'bold'),bg='white')
        lbl_phone.grid(row=1,column=4,padx=2,pady=7,sticky=W)
        
        txt_phone=ttk.Entry(upper_frame,textvariable=self.var_phone,font=('arial',12),width=22)
        txt_phone.grid(row=1,column=5,padx=2,pady=7,sticky=W)
        
         #Label dob (Date Of Birth)
        lbl_dob=Label(upper_frame,text='DOB:',font=('arial',12,'bold'),bg='white')
        lbl_dob.grid(row=2,column=4,padx=2,pady=7,sticky=W)
        
        txt_dob=ttk.Entry(upper_frame,textvariable=self.var_dob,font=('arial',12),width=22)
        txt_dob.grid(row=2,column=5,padx=2,pady=7,sticky=W)
        
        #Label address
        lbl_address=Label(upper_frame,text='Address:',font=('arial',12,'bold'),bg='white')
        lbl_address.grid(row=3,column=4,padx=2,pady=7,sticky=W)
        
        txt_email=ttk.Entry(upper_frame,textvariable=self.var_address,font=('arial',12),width=22)
        txt_email.grid(row=3,column=5,padx=2,pady=7,sticky=W)
        
        #Label seniority (Year)
        lbl_seniority=Label(upper_frame,text='Seniority (years):',font=('arial',12,'bold'),bg='white')
        lbl_seniority.grid(row=4,column=4,padx=2,pady=7,sticky=W)
        
        lbl_seniority=Label(upper_frame,textvariable=self.var_seniority,font=('arial',12,'bold'),bg='white')
        lbl_seniority.grid(row=4,column=5,padx=2,pady=7,sticky=W)
        
        #Label seniority bonus (USD)
        lbl_seniority=Label(upper_frame,text='Experience bonus (USD):',font=('arial',12,'bold'),bg='white')
        lbl_seniority.grid(row=5,column=4,padx=2,pady=7,sticky=W)
        
        lbl_seniority=Label(upper_frame,textvariable=self.var_expbonus,font=('arial',12,'bold'),bg='white')
        lbl_seniority.grid(row=5,column=5,padx=2,pady=7,sticky=W)
        
        #button Frame
        button_frame=Frame(upper_frame,bd=2,relief=RIDGE,bg='white')
        button_frame.place(x=1130,y=5,width=170,height=250)
         
        btn_add=Button(button_frame,text='Save',command=self.add_data,font=('arial',15,'bold'),width=13,bg='blue',fg='white')
        btn_add.grid(row=0,column=0,padx=1,pady=5)
        
        btn_update=Button(button_frame,text='Update',command=self.update_data,font=('arial',15,'bold'),width=13,bg='blue',fg='white')
        btn_update.grid(row=1,column=0,padx=1,pady=5)
        
        btn_delete=Button(button_frame,text='Delete',command=self.delete_data,font=('arial',15,'bold'),width=13,bg='blue',fg='white')
        btn_delete.grid(row=2,column=0,padx=1,pady=5)
        
        btn_clear=Button(button_frame,text='Clear',command=self.reset_data,font=('arial',15,'bold'),width=13,bg='blue',fg='white')
        btn_clear.grid(row=3,column=0,padx=1,pady=5)
        
        btn_seniority=Button(button_frame,text='Seniority',command=self.seniority_bonus,font=('arial',15,'bold'),width=13,bg='blue',fg='white')
        btn_seniority.grid(row=4,column=0,padx=1,pady=5)

        #down Frame
        down_frame=LabelFrame(main_frame,bd=2,relief=RIDGE,text='Employee Information Table',font=('time new roman',13,'bold'),fg='red',bg='white')
        down_frame.place(x=10,y=310,width=1320,height=290)
       
        #Search Frame
        search_frame=LabelFrame(down_frame,bd=2,relief=RIDGE,text='Search Employee Information',font=('time new roman',13,'bold'),fg='red',bg='white')
        search_frame.place(x=10,y=0,width=1295,height=60)
        
        search_by=Label(search_frame,text='Search By:',font=('arial',13,'bold'),fg='red',bg='white')
        search_by.grid(row=0,column=0,sticky=W,padx=5)
        
        #button Search
        self.var_com_search=StringVar()
        com_txt_search=ttk.Combobox(search_frame,textvariable=self.var_com_search,font=('arial',12,'bold'),width=17,state='readonly')
        com_txt_search['value']=('Select Option','id_employee','Name','Phone','id_proof')
        com_txt_search.current(0)
        com_txt_search.grid(row=0,column=1,sticky=W,padx=5)
        
        self.var_search=StringVar()
        txt_search=ttk.Entry(search_frame,textvariable=self.var_search,font=('arial',12),width=22)
        txt_search.grid(row=0,column=2,padx=5)
        
        btn_search=Button(search_frame,text='Search',command=self.search_data,font=('arial',13,'bold'),width=14,bg='Blue',fg='white')
        btn_search.grid(row=0,column=3,padx=5)
        
        btn_search=Button(search_frame,text='Search All',command=self.fetch_data,font=('arial',13,'bold'),width=14,bg='Blue',fg='white')
        btn_search.grid(row=0,column=4,padx=5)
        
#         =============Employee table===================
        # Table Frame
        table_frame=Frame(down_frame,bd=3,relief=RIDGE)
        table_frame.place(x=10,y=60,width=1295,height=200)
        
        scroll_x=ttk.Scrollbar(table_frame,orient=HORIZONTAL)
        scroll_y=ttk.Scrollbar(table_frame,orient=VERTICAL)
        
        self.employee_table= ttk.Treeview(table_frame,column=('idemployee','dep','degi','name','gender','email','phone','dob','doj','married','idproofcomb','idproof',
                                                                'contry','address','salary'),xscrollcommand=scroll_x.set,yscrollcommand=scroll_y.set)
        scroll_x.pack(side=BOTTOM,fill=X)
        scroll_y.pack(side=RIGHT,fill=Y)
        
        scroll_x.config(command=self.employee_table.xview)
        scroll_y.config(command=self.employee_table.yview)
        
        self.employee_table.heading('idemployee', text='ID Employee')
        self.employee_table.heading('dep', text='Department')
        self.employee_table.heading('degi', text='Designition')
        self.employee_table.heading('name', text='Name')
        self.employee_table.heading('gender', text='Gender')
        self.employee_table.heading('email', text='Email')
        self.employee_table.heading('phone', text='Phone')
        self.employee_table.heading('dob', text='DOB')
        self.employee_table.heading('doj', text='DOJ')
        self.employee_table.heading('married', text='Married Status')
        self.employee_table.heading('idproofcomb', text='ID Type')
        self.employee_table.heading('idproof', text='ID Proof')
        self.employee_table.heading('contry', text='Country')
        self.employee_table.heading('address', text='Address')
        self.employee_table.heading('salary', text='Salary')
        
        #co dan cot
        self.employee_table['show']='headings'
        
        self.employee_table.column('idemployee',width=100)
        self.employee_table.column('dep',width=100)
        self.employee_table.column('degi',width=100)
        self.employee_table.column('name',width=100)
        self.employee_table.column('gender',width=100)
        self.employee_table.column('email',width=100)
        self.employee_table.column('phone',width=100)
        self.employee_table.column('dob',width=100)
        self.employee_table.column('doj',width=100)
        self.employee_table.column('married',width=100)
        self.employee_table.column('idproofcomb',width=100)
        self.employee_table.column('idproof',width=100)
        self.employee_table.column('contry',width=100)
        self.employee_table.column('address',width=100)
        self.employee_table.column('salary',width=100)
        
        #Đăng ký tên, gán địa chỉ vào socket
        self.employee_table.pack(fill=BOTH,expand=1)
        self.employee_table.bind('<ButtonRelease>',self.get_cursor)
        self.fetch_data()
        
#       *********************Function Declarations******************
    # add
    def add_data(self):
        if self.var_dep.get()=="Select Department" or self.var_email.get()=="" or self.var_idproof.get()=="":
            messagebox.showerror('Error','All Fields are required')
        else:
            try:
                conn=mysql.connector.connect(host='localhost',user='root',password='Abcd@1234',database='mydata')
                my_cursor=conn.cursor()
                my_cursor.execute('insert into employee2022 values(%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)',(
                                                                                                                self.var_idemployee.get(),
                                                                                                                self.var_dep.get(),
                                                                                                                self.var_designition.get(),
                                                                                                                self.var_name.get(),
                                                                                                                self.var_gender.get(),
                                                                                                                self.var_email.get(),
                                                                                                                self.var_phone.get(),
                                                                                                                self.var_dob.get(),
                                                                                                                self.var_doj.get(),
                                                                                                                self.var_married.get(),
                                                                                                                self.var_idproofcomb.get(),
                                                                                                                self.var_idproof.get(),
                                                                                                                self.var_country.get(),
                                                                                                                self.var_address.get(),
                                                                                                                self.var_salary.get()           
                                                                                                                ))
                conn.commit()
                self.fetch_data()
                conn.close()
                messagebox.showinfo('Success','Employee has been added',parent=self.root)
            except Exception as es:
                messagebox.showerror('Error',f'Due to: {str(es)}',parent=self.root)
            
    # fetch Data
    def fetch_data(self):
        conn=mysql.connector.connect(host='localhost',user='root',password='Abcd@1234',database='mydata')
        my_cursor=conn.cursor()
        my_cursor.execute('select * from employee2022')
        data=my_cursor.fetchall()
        if len(data)!=0:
            self.employee_table.delete(*self.employee_table.get_children())
            for i in data:
                self.employee_table.insert('',END,values=i)
            conn.commit()
        conn.close()
        
    # Get Cursor
    def get_cursor(self,event=''):
        cursor_row=self.employee_table.focus()
        content=self.employee_table.item(cursor_row)
        data=content['values']
        
        self.var_idemployee.set(data[0])
        self.var_dep.set(data[1])
        self.var_designition.set(data[2])
        self.var_name.set(data[3])
        self.var_gender.set(data[4])
        self.var_email.set(data[5])
        self.var_phone.set(data[6])
        self.var_dob.set(data[7])
        self.var_doj.set(data[8])
        self.var_married.set(data[9])
        self.var_idproofcomb.set(data[10])
        self.var_idproof.set(data[11])
        self.var_country.set(data[12])
        self.var_address.set(data[13])
        self.var_salary.set(data[14])
    # Update   
    def update_data(self):
        if self.var_dep.get()=="Select Department" or self.var_email.get()=="" or self.var_idproof.get()=="":
            messagebox.showerror('Error','All Fields are required')
        else:
            try:
                update=messagebox.askyesno('Update','Are you sure update this employee data')
                if update>0:
                    conn=mysql.connector.connect(host='localhost',user='root',password='Abcd@1234',database='mydata')
                    my_cursor=conn.cursor()
                    my_cursor.execute('Update employee2022 set Department=%s,Designition=%s,Name=%s,Gender=%s,Email=%s,Phone=%s,DOB=%s,DOJ=%s,Married_Status=%s,ID_Type=%s,ID_Proof=%s,Country=%s,Address=%s,Salary=%s where ID_Employee=%s',(
                                                                                                                                                                                                                                            
                                                                                                                                                                                                                                            self.var_dep.get(),
                                                                                                                                                                                                                                            self.var_designition.get(),
                                                                                                                                                                                                                                            self.var_name.get(),
                                                                                                                                                                                                                                            self.var_gender.get(),
                                                                                                                                                                                                                                            self.var_email.get(),
                                                                                                                                                                                                                                            self.var_phone.get(),
                                                                                                                                                                                                                                            self.var_dob.get(),
                                                                                                                                                                                                                                            self.var_doj.get(),
                                                                                                                                                                                                                                            self.var_married.get(),
                                                                                                                                                                                                                                            self.var_idproofcomb.get(),
                                                                                                                                                                                                                                            self.var_idproof.get(),
                                                                                                                                                                                                                                            self.var_country.get(),
                                                                                                                                                                                                                                            self.var_address.get(),
                                                                                                                                                                                                                                            self.var_salary.get(),
                        
                                                                                                                                                                                                                                            self.var_idemployee.get()
                                                                                                                                                                                                                                   
                                                                                                                                                                                                                                            ))
                else:
                    if not update:
                        return
                conn.commit()
                self.fetch_data()
                conn.close()
                messagebox.showinfo('success','Employee Successfully update',parent=self.root)
            except Exception as es:
                messagebox.showerror('Error',f'Due to: {str(es)}',parent=self.root)
    # Delete
    def delete_data(self):
        if self.var_idemployee.get()=="":
            messagebox.showerror('Error','All fields are required')
        else:
            try:
                Delete=messagebox.askyesno('Delete','Are you sure delete this employee',parent=self.root)
                if Delete>0:
                    conn=mysql.connector.connect(host='localhost',user='root',password='Abcd@1234',database='mydata')
                    my_cursor=conn.cursor()
                    sql='delete from employee2022 where ID_Employee=%s'
                    value=(self.var_idemployee.get(),)
                    my_cursor.execute(sql,value)
                else:
                    if not Delete:
                        return
                conn.commit()
                self.fetch_data()
                conn.close()
                messagebox.showinfo('Delete','Employee Successfully Delete',parent=self.root)
            except Exception as es:
                messagebox.showerror('Error',f'Due to: {str(es)}',parent=self.root)
    
    # reset
    def reset_data(self):
        self.var_idemployee.set('')
        self.var_dep.set('Select Department')
        self.var_designition.set('')
        self.var_name.set('')
        self.var_gender.set('Male')
        self.var_email.set('')
        self.var_phone.set('')
        self.var_dob.set('')
        self.var_doj.set('')
        self.var_married.set('Married')
        self.var_idproofcomb.set('Select ID Proof')
        self.var_idproof.set('')
        self.var_country.set('')
        self.var_address.set('')
        self.var_salary.set('')
        self.var_seniority.set('')
        self.var_expbonus.set('')
        
    # search
    def search_data(self):
        if self.var_com_search.get()=='' or self.var_search.get()=='':
            messagebox.showerror('Error','Please select option')
        else:
            try:
                conn=mysql.connector.connect(host='localhost',user='root',password='Abcd@1234',database='mydata')
                my_cursor=conn.cursor()
                my_cursor.execute('select * from employee2022 where ' +str(self.var_com_search.get())+" LIKE '%"+str(self.var_search.get()+"%'"))
                rows=my_cursor.fetchall()
                if len(rows)!=0:
                    self.employee_table.delete(*self.employee_table.get_children())
                    for i in rows:
                        self.employee_table.insert('',END,values=i)
                else:
                    messagebox.showerror('Error',f"Sorry, SYSTEM couldn't find any results for '{self.var_search.get()}'")
                conn.commit()
                conn.close()
            except Exception as es:
                messagebox.showerror('Error',f'Due to: {str(es)}',parent=self.root)
    #  Seniority bonus          
    def seniority_bonus(self):
        today = datetime.today()
        date_format = "%d/%m/%Y"
        d0 = today.strftime(date_format)
        d1 = datetime.strptime(self.var_doj.get(), date_format)
        d2 = datetime.strptime(d0, date_format)
        delta = abs(d2 - d1)
        days=delta.days
    #  print(days)
        years=delta.days//365
    #  print(years)
        self.var_seniority.set(f'{years} years')
        if self.var_doj.get()=='' or d1>d1:
            messagebox.showerror('Error','DOJ not correct')
        elif years >= 20:
            self.var_expbonus.set('20.000')
        elif years >= 15:
            self.var_expbonus.set('15.000')
        elif years >=10:
            self.var_expbonus.set('10.000')
        elif years >= 5:
            self.var_expbonus.set('5.000')
        else:
            self.var_expbonus.set('1.000')
# RUN        
if __name__=="__main__":
    root=Tk()
    obj=Employee(root)
    root.mainloop()
```
